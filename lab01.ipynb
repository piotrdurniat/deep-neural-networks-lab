## <span lang="en">Tensory</span>

<span lang="en">Tensory</span> są strukturami danych, które są bardzo podobne do tablic i macierzy. W środowisku **<span lang="en">PyTorch</span>** tensory są wykorzystywane do zapisywania danych wejściowych i wyjściowych modelu, jak również jego parametrów. Tensory są podobne do tablic **<span lang="en">NumPy</span>**, z tą różnicą, że tensory mogą być uruchamiane na kartach graficznych lub innym wyspecjalizowanym sprzęcie w celu przyspieszenia obliczeń. Przed rozpoczęciem prac należy zaimportować bibliotekę **<span lang="en">PyTorch</span>**. Dla przedstawienia pewnych analogii zaimportowana będzie również biblioteka **<span lang="en">NumPy</span>**.

Ustaw kursor na polu z kodem poniżej i wciśnij przycisk Obrazek przedstawiający przycisk uruchamiania kodu z komórki


```python
import torch  # biblioteka PyTorch
import numpy as np  # biblioteka NumPy
```

## Inicjalizacja tensora

Tensory mogą być inicjalizowane na wiele różnych sposobów. Przykładowo:

### Bezpośrednio z listy

Tensory mogą być tworzone bezpośrednio z listy. Typ danych jest określany automatycznie.


```python
data_list = [
    [2, 5],
    [3, 6],
    [4, 7],
]  # lista list z elementami w postaci liczb całkowitych
data_tensor = torch.tensor(data_list)  # utworzenie tensora z listy
data_tensor
```

Utworzony tensor reprezentuje macierz o wymiarach 3x2:

$\begin{bmatrix}
2 & 5 \\
3 & 6 \\
4 & 7
\end{bmatrix}$

Wymiar tensora można sprawdzić następująco:


```python
data_tensor.shape  # kształt tensora
```

Do elementów **<span lang="en">shape</span>** odwołujemy się jak w przypadku listy:


```python
data_tensor.shape[0]
```

```python
data_tensor.shape[1]
```

### Z tablicy <span lang="en">NumPy</span>

Tensory można również tworzyć z tablicy <span lang="en">NumPy</span>


```python
numpy_array = np.array(data_list)  # tablica NumPy utworzona z listy
data_tensor = torch.from_numpy(numpy_array)  # tensor utworzony z tablicy NumPy
data_tensor
```

Tensor można również skonwertować do tablicy <span lang="en">NumPy</span>


```python
np_arr = data_tensor.numpy()  # tablica NumPy utworzona z tensora
np_arr
```

## Operacje na tensorach

Wiele operacji na tensorach w PyTorch jest bardzo podobnych do operacji w NumPy. Niektóre przykłady będą zawierały analogie przedstawione na tablicach <span lang="en">NumPy</span>.


### Indeksowanie

Najpierw utworzone zostaną dwie zmienne: **tensor** przechowująca tensor o wymiarach 3x4 oraz **<span lang="en">array</span>** będącą tablicą NumPy utworzoną na podstawie **tensor**.


```python
tensor = torch.zeros(3, 4)
array = tensor.numpy()
print(tensor)
print(array)
```

### Modyfikacja elementu o podanym indeksie

Indeksy działają analogicznie jak w przypadku [<span lang="en">tablic NumPy</span>](https://numpy.org/doc/stable/reference/arrays.indexing.html).


```python
tensor[1, 2] = 1
array[1, 2] = 1
print(tensor)
print(array)
```

### Wydzielenie fragmentu tensora/tablicy


```python
tensor[:, 2:]
```

```python
array[1:3, 1:4]
```

### Łączenie tensorów


```python
t1 = torch.zeros(3, 2)
t2 = torch.ones(3, 1)
t3 = torch.cat([t1, t2, t1], dim=1)
t3
```

### Mnożenie tensorów


```python
# zdefiniowanie 3 przykładowych tensorów
t1 = torch.tensor([[1, 2], [3, 4], [2, 3]])
t2 = torch.tensor([[2, 5], [3, 6], [4, 7]])
t3 = torch.tensor([[2, 3]])

# mnożenie dwóch tensorów (Uwaga! To nie jest mnożenie macierzowe znane z algebry!)
t4 = t1.mul(t2)
t5 = t1.mul(t3)

# mnożenie macierzy przez liczbę
t6 = t1.mul(5)
print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
print(t6)
```

```python
# alternatywnie dla powyższych operatorów, można zapisać:
t4 = t1 * t2
t5 = t1 * t3
t6 = t1 * 5
print(t4)
print(t5)
print(t6)
```

```python
# transpozycja macierzy
print(t1)
print(t1.T)
```

### Operacje "w miejscu" (<span lang="en">in-place</span>)


```python
# Uwaga! Operacje w miejscu pomagają oszczędzić pamięć, ale mogą nadpisać wartości potrzebne do wyznaczania gradientu.
# Więcej informacji można znaleźć tu: https://pytorch.org/docs/stable/notes/autograd.html#in-place-operations-with-autograd

# Operacje w miejscu są zdefiniowane z sufiksem "_"
print(t1)
t1.add_(5)
print(t1)
t1.mul_(2)
print(t1)
```
